```python
query = ('select * from pedidos')
pedidos_df = pd.read_sql(query, engine)
pedidos_df
```

```python
# prompt: Usando o DataFrame pedidos_df: sazonal

import pandas as pd
# Convert the order_purchase_timestamp column to datetime objects
pedidos_df['order_purchase_timestamp'] = pd.to_datetime(pedidos_df['order_purchase_timestamp'])
# Extract the month from the timestamp and create a new column
pedidos_df['month'] = pedidos_df['order_purchase_timestamp'].dt.month
# Group by month and count the number of orders
seasonal_orders = pedidos_df.groupby('month').size().reset_index(name='order_count')
# Print the resulting dataframe
seasonal_orders
```
